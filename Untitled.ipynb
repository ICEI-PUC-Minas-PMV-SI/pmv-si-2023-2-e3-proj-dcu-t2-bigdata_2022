```python
import json
import pandas as pd
from tqdm import tqdm

import seaborn as sns
import matplotlib.pyplot as plt

from pygments import highlight
from pygments.lexers import JsonLexer
from pygments.formatters import TerminalFormatter

from google_play_scraper import Sort, reviews, app

%matplotlib inline
%config InlineBackend.figure_format='retina'

sns.set(style='whitegrid', palette='muted', font_scale=1.2)
```

```python
# Lista de 9 aplicações para analisar e extrair informações básicas sobre cada um deles.
apps_ids = ['br.com.brainweb.ifood'
            #,'com.cerveceriamodelo.modelonow'
            #,'com.mcdo.mcdonalds'
            #,'habibs.alphacode.com.br'
            ,'com.xiaojukeji.didi.brazil.customer'
            ,'com.ubercab.eats'
            ,'com.grability.rappi'
            #,'burgerking.com.br.appandroid'
            ]
```

```python
# Coletando os dados

app_infos = []

for ap in tqdm(apps_ids):
    info = app(ap, lang='en', country='us')
    del info['comments']
    app_infos.append(info)
```

```python

app_infos_df = pd.DataFrame(app_infos)
app_infos_df
```

```python
# Coletei as avaliações dos usuários para cada um dos aplicativos. 
# Como os textos serão divididos em três sentimentos possíveis (positivo, negativo ou neutro) 
# o objetivo é obter um dataset o mais balanceado possível, selecionei 4000 amostras de avaliações com score 3 e 2000 de cada uma das outras.
    
app_reviews = []

for ap in tqdm(apps_ids):
    for score in list(range(1, 6)):
        for sort_order in [Sort.MOST_RELEVANT, Sort.NEWEST]:
            rvs, _ = reviews(
                ap,
                lang='pt',
                country='br',
                sort=sort_order,
                count= 4000 if score == 3 else 2000,
                filter_score_with=score
            )
            for r in rvs:
                r['sortOrder'] = 'most_relevant' if sort_order == Sort.MOST_RELEVANT else 'newest'
                r['appId'] = ap
            app_reviews.extend(rvs)
```

```python
df = pd.DataFrame(app_reviews)
```

```python
# Distribuição da variavel sentiment
colors = sns.color_palette("pastel") # deep, pastel, Set1 Set2 Set3, icefire, tab10, muted, colorlind, coolwarm

col = 'score'

total = len(df)

plt.figure(figsize=(12,6))
g = sns.countplot(x=col, data=df, palette=colors)
g.set_title(f"Distribuição da variável {col}")
g.set_xlabel(f"{col}")
g.set_ylabel("Quantidade")
sizes=[]
for p in g.patches:
    height = p.get_height()
    sizes.append(height)
    g.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format((height/total)*100),
            ha="center", fontsize=14) 
g.set_ylim(0, max(sizes) * 1.15)
plt.show()
```

```python
# A ideia é que sejam atribuídos sentimentos para avaliações a partir dos scores seguindo a seguinte divisão:
# 1,2 = 0 (Negativo)
# 3 = 2 (Neutro)
# 4,5 = 1 (Positivo)

def to_sentiment(rating):
  rating = int(rating)
  if rating <= 2:
    return 0
  elif rating == 3:
    return 2
  else: 
    return 1

def to_query_used(rating):
  rating = int(rating)
  if rating <= 2:
    return ':('
  elif rating == 3:
    return ':|'
  else: 
    return ':)'

df['sentiment'] = df.score.apply(to_sentiment)
df['query_used'] = df.score.apply(to_query_used)
```

```python
# Distribuição da variavel sentiment

col = 'sentiment'

total = len(df)

plt.figure(figsize=(12,6))
g = sns.countplot(x=col, data=df, palette=colors)
g.set_title(f"Distribuição da variável {col}")
g.set_xlabel(f"{col}")
g.set_ylabel("Quantidade")
sizes=[]
for p in g.patches:
    height = p.get_height()
    sizes.append(height)
    g.text(p.get_x()+p.get_width()/2.,
            height + 3,
            '{:1.2f}%'.format((height/total)*100),
            ha="center", fontsize=14) 
g.set_ylim(0, max(sizes) * 1.15)
plt.show()
```
